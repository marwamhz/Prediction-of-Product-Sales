```python
import pandas as pd
```

```python
# load data
df = pd.read_csv("Data/sales_predictions_2023.csv")
df.head()
```

```python
# Find Correct inconsistencies in the Item_Fat_Content column
df['Item_Fat_Content'].value_counts()
```

```python
# Correct inconsistencies in the Item_Fat_Content column
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'})
# check the value counts for Item_Fat_Content column
df['Item_Fat_Content'].value_counts()
```

```python
# Drop Item_Identifer, Outlet_Identifier, and Outlet_Establishment_Year
df = df.drop(columns = ['Item_Identifier', 'Outlet_Identifier', 'Outlet_Establishment_Year'])
df.head()
```

```python
# saving the dataframe
df.to_csv('Data/sales_predictions_2023_modified.csv', index=False)
```
